```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z1 = 1-5j
z2 = 3+2j
z3 = -7+1j
z4 = -1.6j
```

```python
def plot_complex(c1, c2, c3, c4):
    fig, ax = plt.subplots(figsize=(5, 5))

    # Extract real and imaginary parts of z1 and z2
    x1, y1 = z1.real, z1.imag
    x2, y2 = z2.real, z2.imag
    x3, y3 = z3.real, z3.imag
    x4, y4 = z4.real, z4.imag

    ax.set_title('Plot of z1-z4')

    ax.set_xlabel('Re (Real Part)')
    ax.set_ylabel('Im (Imaginary Part)')

    # Set appropriate limits for the plot
    ax.set_xlim((-10, 10))
    ax.set_ylim((-10, 10))

    ax.axhline(color='k')
    ax.axvline(color='k')

    # Plot the first complex number
    if y1 > 0:
        ax.vlines(x1, 0, y1, 'gray', ':')
    else:
        ax.vlines(x1, y1, 0, 'gray', ':')

    if x1 > 0:
        ax.hlines(y1, 0, x1, 'gray', ':')
    else:
        ax.hlines(y1, x1, 0, 'gray', ':')

    ax.text(x1, y1 / 2, f'({y1})')
    ax.text(x1 / 2 - 0.3, y1 - 0.3, f'({x1})')
    ax.plot(x1, y1, 'o', label="z1")

    # Plot the second complex number
    if y2 > 0:
        ax.vlines(x2, 0, y2, 'gray', ':')
    else:
        ax.vlines(x2, y2, 0, 'gray', ':')

    if x2 > 0:
        ax.hlines(y2, 0, x2, 'gray', ':')
    else:
        ax.hlines(y2, x2, 0, 'gray', ':')

    ax.text(x2, y2 / 2, f'({y2})')
    ax.text(x2 / 2 - 0.3, y2 - 0.3, f'({x2})')
    ax.plot(x2, y2, 'o', label="z2")

    # Plot the third complex number
    if y3 > 0:
        ax.vlines(x3, 0, y3, 'gray', ':')
    else:
        ax.vlines(x3, y3, 0, 'gray', ':')

    if x3 > 0:
        ax.hlines(y3, 0, x3, 'gray', ':')
    else:
        ax.hlines(y3, x3, 0, 'gray', ':')

    ax.text(x3, y3 / 2, f'({y3})')
    ax.text(x3 / 2 - 0.3, y3 - 0.3, f'({x3})')
    ax.plot(x3, y3, 'o', label="z3")

    # Plot the fourth complex number
    if y4 > 0:
        ax.vlines(x4, 0, y4, 'gray', ':')
    else:
        ax.vlines(x4, y4, 0, 'gray', ':')

    if x4 > 0:
        ax.hlines(y4, 0, x4, 'gray', ':')
    else:
        ax.hlines(y4, x4, 0, 'gray', ':')

    ax.text(x4, y4 / 2, f'({y4})')
    ax.text(x4 / 2 - 0.3, y4 - 0.3, f'({x4})')
    ax.plot(x4, y4, 'o', label="z4")

    ax.legend()

    ax.grid()
```

```python
plot_complex(z1,z2,z3,z4)
```
