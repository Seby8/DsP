```python
import numpy as np
import matplotlib.pyplot as plt
```

##Part 1##

```python
h = np.array([
    [1,1],
    [1,-1]
])

H = np.dot(1/np.sqrt(2),h)

y = np.array([
    [0,1],
    [-1,0]
])

o = np.array([
    [1,0],
    [1,1]
])
```

```python
#Getting the Magnitude
def v_norm(n, vector):
    n_sum = np.sum(vector**n)
    vect_norm = np.power(n_sum,1/n)
    return vect_norm

def l1_norm(vector):
    vect_norm = np.sum(vector)
    return vect_norm
```

```python
#1
dot_1 = np.dot(H,y)
print(f"Magnitude: {v_norm(1, dot_1)}")

angle_rad = np.arctan2(dot_1[1, 0], dot_1[0, 0])
angle_deg = np.degrees(angle_rad)

print(f"Angle: {angle_deg}")
```

```python
plt.plot(dot_1, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Plot')
plt.grid()
plt.show()
```

```python
#2
dot_2 = np.dot(y,H)
print(f"Magnitude: {v_norm(2, dot_2)}")

angle_rad = np.arctan2(dot_2[1, 0], dot_2[0, 0])
angle_deg = np.degrees(angle_rad)

print(f"Angle: {angle_deg}")
```

```python
plt.plot(dot_2, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Plot')
plt.grid()
plt.show()
```

```python
#3
dot_3 = np.dot(H,H)
print(f"Magnitude: {v_norm(2, dot_3)}")

angle_rad = np.arctan2(dot_3[1, 0], dot_3[0, 0])
angle_deg = np.degrees(angle_rad)

print(f"Angle: {angle_deg}")
```

```python
plt.plot(dot_3, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Plot')
plt.grid()
plt.show()
```

```python
#4
dot_4 = y.dot(H.dot(o))
print(f"Magnitude: {v_norm(2, dot_4)}")

angle_rad = np.arctan2(dot_4[1, 0], dot_4[0, 0])
angle_deg = np.degrees(angle_rad)

print(f"Angle: {angle_deg}")
```

```python
#5
dot_5 = H.dot(y.dot(H.dot(o)))
print(f"Magnitude: {v_norm(2, dot_5)}")

angle_rad = np.arctan2(dot_5[1, 0], dot_5[0, 0])
angle_deg = np.degrees(angle_rad)

print(f"Angle: {angle_deg}")
```

##Part 3##

```python
a = np.array([
    [5,0,0],
    [0,5,0],
    [0,0,5]
])

b = np.array([
    [1,0,5],
    [2,7,6],
    [6,4,7]
])

c = np.array([
    [1,2,6],
    [3,15,4],
    [2,10,3]
])

d = np.array([
    [5,2,4],
    [6,2,4],
    [0,1,1]
])
```

```python
#1
dot_ab = np.dot(a,b)
det_1 = np.linalg.det(dot_ab)

if(det_1 != 0).any():
  print("Linearly depdent")
  print(det_1)
else:
  print("Not linearly dependent")
```

```python
#2
dot_cd = np.dot(c,d)
det_2 = np.linalg.det(dot_cd)

if(det_2 != 0).any():
  print("Linearly depdent")
  print(det_2)
else:
  print("Not linearly dependent")
```

```python
time = np.array([0, np.pi/4,
              np.pi/2, 3*np.pi/4,
              np.pi])
corr = np.array([5, 3,
              0, -3,
              5])

sig = corr * np.sin(time)

plt.plot(time, sig, marker='o')
plt.xlabel(f"Time")
plt.ylabel(f"Amplitude")
plt.title(f"Signal Corresponding")
plt.grid()
plt.show()
```
